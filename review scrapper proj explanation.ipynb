```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipkart_url1 = "https://www.flipkart.com/search?q=samsung"
```

```python
uClient = uReq(flipkart_url1)

# url open from urllib.request imported as uReq above and it is used to siply open a url.
uClient

#If we pass wrong url, it gives error.
```

```python
flipkartPage = uClient.read()
flipkartPage #read text

# But not returned in a very readable format, when we rightclick on any page and click on view page source, it prints
# the same below format of any webpage.
```

```python
uClient.close()
```

```python
flipkart_html = bs(flipkartPage, "html.parser")# bs is beautiful soap library which beautifies the unreadable text.
flipkart_html
#This output and above both are almost same but the below one is just structured line by line.
```

now to filter out only those boxes which are search results of iphone or any such product, click fn+F12 which shows html tags of any webpage, then click on top right mouse cursor of that page which changes mouse cursor so as wherever we click on main
screen, it's corresponding tag containing it's division class appears, their may be subclasses also inside one main  div class, but by scrolling mouse on different boxes of product search results, 1st scroll mouse in such a way that it is able to select the whole product result box, we get different tags containing same division class e.g. in below case for iphone search results which represents each product result box and in below iphone e.g. name of that division class is _1AtVbE col-12-12.

searching that division class out of whole flipkart_html prints all iphone based results and len of that = 15 means there
have been 15 iphone search results.

```python
bigBoxes = flipkart_html.find_all("div",{"class":"_1AtVbE col-12-12"})
bigBoxes
# We are searching for division class named "_1AtVbE col-12-12"
```

Now out of this url page, we need to go to reviews section

```python
prodRes = requests.get(productLink)
prodRes
```

```python
prodRes.encoding='utf-8' #encode it in english format utf-8 and nothing else.
prod_html = bs(prodRes.text, "html.parser")#beautifulsoup for better presentable as done above
print(prod_html)
```

Now when we go through the above url which opens product, we again have to search for div class for boxes of all reviews of that product as we did above and class turns out to be "_16PBlm".

```python
commentboxes = prod_html.find_all('div', {'class': "_16PBlm"})
commentboxes
```
